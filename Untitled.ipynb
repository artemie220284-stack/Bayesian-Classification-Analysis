```python
class LaplaceNaiveBayes:
    """拉普拉斯修正的朴素贝叶斯分类器（混合特征）"""
    
    def __init__(self):
        self.classes_ = None
        self.priors_ = None  # 先验概率
        self.discrete_likelihoods_ = {}  # 离散特征的条件概率
        self.continuous_params_ = {}  # 连续特征的参数（均值和方差）
        self.feature_values_ = {}  # 各离散特征的取值集合
        
    def fit(self, X_discrete, X_continuous, y):
        """训练模型"""
        self.classes_ = np.unique(y)
        n_samples = len(y)
        
        # 1. 计算先验概率（也使用拉普拉斯修正）
        for c in self.classes_:
            count = np.sum(y == c)
            self.priors_ = {} if self.priors_ is None else self.priors_
            # 拉普拉斯修正：分子+1，分母+类别数
            self.priors_[c] = (count + 1) / (n_samples + len(self.classes_))
        
        # 2. 收集每个离散特征的取值集合
        n_discrete_features = X_discrete.shape[1]
        for i in range(n_discrete_features):
            self.feature_values_[i] = np.unique(X_discrete[:, i])
        
        # 3. 计算离散特征的条件概率（使用拉普拉斯修正）
        self.discrete_likelihoods_ = {}
        
        for c in self.classes_:
            class_mask = (y == c)
            class_samples = X_discrete[class_mask]
            n_class = len(class_samples)
            
            self.discrete_likelihoods_[c] = {}
            
            for i in range(n_discrete_features):
                self.discrete_likelihoods_[c][i] = {}
                k_i = len(self.feature_values_[i])  # 第i个特征的取值个数
                
                for value in self.feature_values_[i]:
                    # 统计类别c中特征i取值为value的样本数
                    count = np.sum(class_samples[:, i] == value)
                    # 拉普拉斯修正
                    prob = (count + 1) / (n_class + k_i)
                    self.discrete_likelihoods_[c][i][value] = prob
        
        # 4. 计算连续特征的参数（假设服从高斯分布）
        self.continuous_params_ = {}
        
        for c in self.classes_:
            class_mask = (y == c)
            class_samples = X_continuous[class_mask]
            
            self.continuous_params_[c] = {
                'means': np.mean(class_samples, axis=0),
                'stds': np.std(class_samples, axis=0) + 1e-9  # 防止零方差
            }
        
        print("训练完成！")
        print(f"先验概率: {self.priors_}")
        return self
    
    def _gaussian_prob(self, x, mean, std):
        """计算高斯分布概率密度"""
        exponent = np.exp(-((x - mean) ** 2) / (2 * std ** 2))
        return (1 / (np.sqrt(2 * np.pi) * std)) * exponent
    
    def predict_proba(self, X_discrete, X_continuous):
        """预测概率"""
        n_samples = len(X_discrete)
        probabilities = np.zeros((n_samples, len(self.classes_)))
        
        for idx, c in enumerate(self.classes_):
            # 初始化为先验概率的对数
            probs = np.full(n_samples, np.log(self.priors_[c]))
            
            # 乘以离散特征的条件概率（对数相加）
            for i in range(X_discrete.shape[1]):
                for sample_idx in range(n_samples):
                    value = X_discrete[sample_idx, i]
                    if value in self.discrete_likelihoods_[c][i]:
                        prob = self.discrete_likelihoods_[c][i][value]
                        probs[sample_idx] += np.log(prob)
                    else:
                        # 如果遇到训练时未出现的特征值，使用最小概率
                        k_i = len(self.feature_values_[i])
                        min_prob = 1 / (len(X_discrete) + k_i)  # 拉普拉斯修正的最小值
                        probs[sample_idx] += np.log(min_prob)
            
            # 乘以连续特征的概率密度（假设独立）
            if len(self.continuous_params_) > 0:
                for j in range(X_continuous.shape[1]):
                    mean = self.continuous_params_[c]['means'][j]
                    std = self.continuous_params_[c]['stds'][j]
                    
                    for sample_idx in range(n_samples):
                        x = X_continuous[sample_idx, j]
                        # 使用对数概率密度
                        gaussian_log_prob = np.log(self._gaussian_prob(x, mean, std))
                        probs[sample_idx] += gaussian_log_prob
            
            # 存储概率（还是对数形式）
            probabilities[:, idx] = probs
        
        # 转换回概率（使用softmax处理对数概率）
        # 防止数值下溢：减去最大值
        max_log_probs = np.max(probabilities, axis=1, keepdims=True)
        exp_probs = np.exp(probabilities - max_log_probs)
        probabilities_normalized = exp_probs / np.sum(exp_probs, axis=1, keepdims=True)
        
        return probabilities_normalized
    
    def predict(self, X_discrete, X_continuous):
        """预测类别"""
        probabilities = self.predict_proba(X_discrete, X_continuous)
        return self.classes_[np.argmax(probabilities, axis=1)]
    
    def score(self, X_discrete, X_continuous, y):
        """计算准确率"""
        predictions = self.predict(X_discrete, X_continuous)
        return np.mean(predictions == y)
    
    def get_feature_importance(self):
        """获取特征重要性（信息增益）"""
        # 简化的特征重要性分析
        importance = {}
        
        for i, feature_name in enumerate(feature_names[:6]):  # 只分析离散特征
            # 计算该特征在不同类别下的概率差异
            if 0 in self.discrete_likelihoods_ and 1 in self.discrete_likelihoods_:
                diff_sum = 0
                for value in self.feature_values_[i]:
                    prob_0 = self.discrete_likelihoods_[0][i].get(value, 0)
                    prob_1 = self.discrete_likelihoods_[1][i].get(value, 0)
                    diff_sum += abs(prob_0 - prob_1)
                importance[feature_name] = diff_sum / len(self.feature_values_[i])
        
        return importance

# 创建并训练模型
model = LaplaceNaiveBayes()
model.fit(X_discrete, X_continuous, y)
```

```python
def print_model_details(model):
    """打印模型学到的概率"""
    print("=" * 60)
    print("拉普拉斯贝叶斯模型细节")
    print("=" * 60)
    
    print("\n1. 先验概率:")
    for c, prob in model.priors_.items():
        label = "好瓜" if c == 1 else "坏瓜"
        print(f"   P({label}) = {prob:.4f}")
    
    print("\n2. 离散特征条件概率（示例）:")
    for c in model.classes_:
        label = "好瓜" if c == 1 else "坏瓜"
        print(f"\n   === 类别: {label} ===")
        
        for i, feature_name in enumerate(feature_names[:3]):  # 只显示前3个特征
            print(f"\n   {feature_name}:")
            for value, prob in model.discrete_likelihoods_[c][i].items():
                # 原始概率（无平滑）作为参考
                class_mask = (y == c)
                class_samples = X_discrete[class_mask]
                n_class = len(class_samples)
                raw_count = np.sum(class_samples[:, i] == value)
                raw_prob = raw_count / n_class if n_class > 0 else 0
                
                print(f"     {value}: 原始={raw_prob:.3f}, 平滑后={prob:.3f}")
    
    print("\n3. 连续特征参数:")
    for c in model.classes_:
        label = "好瓜" if c == 1 else "坏瓜"
        print(f"\n   {label}:")
        print(f"     密度: μ={model.continuous_params_[c]['means'][0]:.3f}, "
              f"σ={model.continuous_params_[c]['stds'][0]:.3f}")
        print(f"     含糖率: μ={model.continuous_params_[c]['means'][1]:.3f}, "
              f"σ={model.continuous_params_[c]['stds'][1]:.3f}")
    
    print("\n4. 特征重要性分析:")
    importance = model.get_feature_importance()
    for feature, score in sorted(importance.items(), key=lambda x: x[1], reverse=True):
        print(f"   {feature}: {score:.4f}")

print_model_details(model)
```

```python
# 在训练集上测试
train_predictions = model.predict(X_discrete, X_continuous)
train_proba = model.predict_proba(X_discrete, X_continuous)

print("\n" + "=" * 60)
print("训练集预测结果")
print("=" * 60)

for i in range(len(X_discrete)):
    true_label = "好瓜" if y[i] == 1 else "坏瓜"
    pred_label = "好瓜" if train_predictions[i] == 1 else "坏瓜"
    
    prob_good = train_proba[i, 1]  # 好瓜的概率
    prob_bad = train_proba[i, 0]   # 坏瓜的概率
    
    correct = "✓" if train_predictions[i] == y[i] else "✗"
    
    print(f"样本{i+1:2d}: 真实={true_label}, 预测={pred_label} {correct}")
    print(f"      概率: 好瓜={prob_good:.3f}, 坏瓜={prob_bad:.3f}")
    
    # 显示关键特征
    if i < 3:  # 只显示前3个样本的详细信息
        print(f"      特征: 色泽={X_discrete[i,0]}, 密度={X_continuous[i,0]:.3f}")
    print()

# 计算准确率
accuracy = model.score(X_discrete, X_continuous, y)
print(f"训练集准确率: {accuracy*100:.2f}%")
```

```python
class NaiveBayesNoSmoothing:
    """无平滑的朴素贝叶斯（用于对比）"""
    
    def __init__(self):
        self.classes_ = None
        self.priors_ = None
        self.discrete_likelihoods_ = {}
        self.continuous_params_ = {}
        
    def fit(self, X_discrete, X_continuous, y):
        self.classes_ = np.unique(y)
        n_samples = len(y)
        
        # 先验概率（无平滑）
        for c in self.classes_:
            count = np.sum(y == c)
            self.priors_ = {} if self.priors_ is None else self.priors_
            self.priors_[c] = count / n_samples
        
        # 离散特征条件概率（无平滑）
        for c in self.classes_:
            class_mask = (y == c)
            class_samples = X_discrete[class_mask]
            n_class = len(class_samples)
            
            self.discrete_likelihoods_[c] = {}
            
            for i in range(X_discrete.shape[1]):
                self.discrete_likelihoods_[c][i] = {}
                unique_values = np.unique(X_discrete[:, i])
                
                for value in unique_values:
                    count = np.sum(class_samples[:, i] == value)
                    prob = count / n_class if n_class > 0 else 0
                    self.discrete_likelihoods_[c][i][value] = prob
        
        # 连续特征参数
        for c in self.classes_:
            class_mask = (y == c)
            class_samples = X_continuous[class_mask]
            
            self.continuous_params_[c] = {
                'means': np.mean(class_samples, axis=0),
                'stds': np.std(class_samples, axis=0) + 1e-9
            }
        
        return self
    
    def _gaussian_prob(self, x, mean, std):
        exponent = np.exp(-((x - mean) ** 2) / (2 * std ** 2))
        return (1 / (np.sqrt(2 * np.pi) * std)) * exponent
    
    def predict_proba(self, X_discrete, X_continuous):
        n_samples = len(X_discrete)
        probabilities = np.zeros((n_samples, len(self.classes_)))
        
        for idx, c in enumerate(self.classes_):
            probs = np.full(n_samples, np.log(self.priors_[c]))
            
            for i in range(X_discrete.shape[1]):
                for sample_idx in range(n_samples):
                    value = X_discrete[sample_idx, i]
                    if value in self.discrete_likelihoods_[c][i]:
                        prob = self.discrete_likelihoods_[c][i][value]
                        if prob > 0:
                            probs[sample_idx] += np.log(prob)
                        else:
                            # 零概率问题！
                            probs[sample_idx] = -np.inf  # 负无穷
            
            for j in range(X_continuous.shape[1]):
                mean = self.continuous_params_[c]['means'][j]
                std = self.continuous_params_[c]['stds'][j]
                
                for sample_idx in range(n_samples):
                    x = X_continuous[sample_idx, j]
                    gaussian_log_prob = np.log(self._gaussian_prob(x, mean, std))
                    probs[sample_idx] += gaussian_log_prob
            
            probabilities[:, idx] = probs
        
        # 处理无穷大（零概率）的情况
        max_log_probs = np.max(probabilities, axis=1, keepdims=True)
        exp_probs = np.exp(probabilities - max_log_probs)
        
        # 如果有全为0的行（因为-inf），设为均匀分布
        row_sums = np.sum(exp_probs, axis=1)
        zero_mask = row_sums == 0
        if np.any(zero_mask):
            exp_probs[zero_mask] = 1 / len(self.classes_)
            row_sums[zero_mask] = 1
        
        probabilities_normalized = exp_probs / row_sums[:, np.newaxis]
        
        return probabilities_normalized
    
    def predict(self, X_discrete, X_continuous):
        probabilities = self.predict_proba(X_discrete, X_continuous)
        return self.classes_[np.argmax(probabilities, axis=1)]
    
    def score(self, X_discrete, X_continuous, y):
        predictions = self.predict(X_discrete, X_continuous)
        return np.mean(predictions == y)

# 对比两种模型
print("\n" + "=" * 60)
print("拉普拉斯修正 vs 无平滑 对比")
print("=" * 60)

# 训练无平滑模型
model_no_smooth = NaiveBayesNoSmoothing()
model_no_smooth.fit(X_discrete, X_continuous, y)

# 在训练集上比较
laplace_acc = model.score(X_discrete, X_continuous, y)
no_smooth_acc = model_no_smooth.score(X_discrete, X_continuous, y)

print(f"拉普拉斯修正模型准确率: {laplace_acc*100:.2f}%")
print(f"无平滑模型准确率: {no_smooth_acc*100:.2f}%")

# 测试包含未出现特征的样本
test_cases = [
    (['浅白', '硬挺', '清脆', '清晰', '凹陷', '硬滑'], [0.500, 0.300], "包含未出现特征"),
    (['青绿', '蜷缩', '浊响', '清晰', '凹陷', '硬滑'], [0.697, 0.460], "训练集中有的样本"),
]

print("\n样本测试对比:")
for i, (discrete_vals, continuous_vals, description) in enumerate(test_cases):
    discrete = np.array([discrete_vals])
    continuous = np.array([continuous_vals])
    
    pred_laplace = model.predict(discrete, continuous)[0]
    prob_laplace = model.predict_proba(discrete, continuous)[0]
    
    pred_no_smooth = model_no_smooth.predict(discrete, continuous)[0]
    prob_no_smooth = model_no_smooth.predict_proba(discrete, continuous)[0]
    
    print(f"\n测试{i+1} ({description}):")
    print(f"  拉普拉斯修正: 预测={'好瓜' if pred_laplace==1 else '坏瓜'}, "
          f"概率=[好瓜={prob_laplace[1]:.4f}, 坏瓜={prob_laplace[0]:.4f}]")
    print(f"  无平滑: 预测={'好瓜' if pred_no_smooth==1 else '坏瓜'}, "
          f"概率=[好瓜={prob_no_smooth[1]:.4f}, 坏瓜={prob_no_smooth[0]:.4f}]")
```
